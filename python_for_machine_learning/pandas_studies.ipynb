# **Tasks for the Titanic dataset**

```python
import seaborn as sns
import pandas as pd
titanic_df = sns.load_dataset('titanic') # Task-1: Define the Titanic dataset from within the Seaborn library
print(titanic_df['sex'].value_counts()) # Task-2: Find the number of male and female passengers in the Titanic data set
```

```python
print(titanic_df.nunique()) # Task-3: Find the number of unique values for each column
```

```python
print(titanic_df['pclass'].nunique()) # Task-4: Find the number of unique values for pclass column
```

```python
print(titanic_df[['pclass','parch']].nunique()) # Task-5: Find the number of unique values for pclass and parch column
```

```python
# Task-6: Check the type of the embarked column. Change the type of the embarked column to Category and check that again.
print(titanic_df['embarked'].dtype)
titanic_df['embarked'] = titanic_df['embarked'].astype('category')
print(titanic_df['embarked'].dtype)
```

```python
display(titanic_df[titanic_df['embarked']=='C']) # Task-7: Show all information for those with embarked value c.
```

```python
display(titanic_df[titanic_df['embarked']!='S']) # Task-8: Show all information for those with no embarked value S.
```

```python
display(titanic_df[(titanic_df['age']<30) & (titanic_df['sex']=='female')]) # Task-9: Show all information for female passengers under 30 years of age.
```

```python
display(titanic_df[(titanic_df['fare']>500) | (titanic_df['age']>70)]) # Task-10: Show information for passengers with a fare value greater than 500 and age greater than 70 years.
```

```python
print(titanic_df.isnull().sum()) # Task-11: Find the sum of the null values in each variable.
```

```python
titanic_df.drop('who', axis='columns', inplace=True) # Task-12: Remove who variable from the data set.
display(titanic_df.head())
```

```python
titanic_df['deck'].fillna(titanic_df['deck'].mode()[0], inplace=True) # Task-13: Fill the empty values in the deck variable with the most repeated value (mod) of the deck variable.
display(titanic_df.head())
```

```python
titanic_df['age'].fillna(titanic_df['age'].median(), inplace=True) # Task-14: Fill the empty values in the age variable with the median value of the age variable.
display(titanic_df.head())
```

```python
display(titanic_df.groupby(['pclass','sex']).agg({'survived':['sum','count','mean']})) # Task-15: Find the sum, count and mean values of the survived variable by grouping Pclass and gender variables
```

```python
# Task-16: Write a function that gives 1 to those under 30 and 0 to those above and equal to 30.
# Create a variable named age_flag in the titanic dataset using the function you wrote (use apply and lambda constructs).
titanic_df['age_flag'] = titanic_df['age'].apply(lambda x: 1 if x<30 else 0)
titanic_df.head()
```

# **Tasks for the Tips dataset**

```python
tip_df = sns.load_dataset('tips') # Task-17: Define the tips data set from the Seaborn library.
display(tip_df.head())
```

```python
# Task-18: Find the sum, min, max and average values of the total_bill values according to the categories (Dinner, Lunch) of the variable Time.
display(tip_df.groupby('time').agg({'total_bill':['min','max','mean']}))
```

```python
# Task-19: Find the sum, min, max and average of total_bill values according to day and time variables.
display(tip_df.groupby(['day','time']).agg({'total_bill':['min','max','mean']}))
```

```python
# Task-20: Find the sum, min, max and average values of total bill and tip values of female customers at lunch time for each day
tip_df.head()
tip_df[(tip_df['time']=="Lunch") & (tip_df['sex'] == "Female")].groupby(['day']).agg({'total_bill':['min','max','mean'],
                                                                             'tip':['min','max','mean']})
```

```python
# Task-21: Use loc to find the average of orders with a size value less than 3 and a total_bill value greater than 10.
tip_df.loc[(tip_df['size']<3) & (tip_df['total_bill']>10), :].mean(numeric_only=True)
```

```python
# Task-22: Create a new variable called total_bill_tip_sum and assign it the average of the total bill and tip paid by each customer.
tip_df['total_bill_tip_sum'] = tip_df['total_bill']+tip_df['tip']
tip_df.head()
```

```python
# Task-23: Sort the data set from largest to smallest according to the total_bill_tip_sum variable and assign the first 30 rows to a new data set.
new_titanic_df = tip_df.sort_values(by="total_bill_tip_sum", ascending=False)[:30]
new_titanic_df
```
